Part 2: Calculate and plot the regression line.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
age = [3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17]
wing_length = [1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0]

# Perform linear regression
coefficients = np.polyfit(age, wing_length, 1)
slope = coefficients[0]
intercept = coefficients[1]

# Create the regression line
regression_line = [slope * x + intercept for x in age]

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(age, wing_length, marker='o', color='b', label='Wing Length')

# Plot the regression line
plt.plot(age, regression_line, color='r', linestyle='--', label=f'Regression Line (y = {slope:.2f}x + {intercept:.2f})')

# Add labels and a legend
plt.xlabel('Age')
plt.ylabel('Wing Length (cm)')
plt.title('Age vs. Wing Length with Regression Line')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
